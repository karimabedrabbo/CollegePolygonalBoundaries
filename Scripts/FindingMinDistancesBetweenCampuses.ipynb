```python
min_distances = []
min_dist_dict = []
maini = 1
with open(r'C:\Users\Karim\Desktop\campus_v2.csv') as csv_file1:
    csv_reader1 = csv.reader(csv_file1, delimiter=',')
    for urow in csv_reader1:
        if maini % 100 == 0:
            print("at" + str(maini))
        distances = []
        dist_dict = []
        for uindex, uelement in enumerate(urow):
            if uelement is '.':
                print("nolonlat at " + str(maini))
                break
            if uindex == 14:
                ulon = float(uelement)
            if uindex == 15:     
                ulat = float(uelement)
                with open(r'C:\Users\Karim\Desktop\campus_v2.csv') as csv_file2:
                    csv_reader2 = csv.reader(csv_file2, delimiter=',')
                    for lrow in csv_reader2:
                        for lindex, lelement in enumerate(lrow):
                            if lelement is '.':
                                break
                            if lindex == 14:
                                llon = float(lelement)
                            if lindex == 15:
                                llat = float(lelement)
                                distance = mpu.haversine_distance((llat, llon), (ulat, ulon))
                                distances.append(distance)
                                dist_dict.append(({urow[1]: (ulat, ulon)}, {lrow[1]: (llat, llon)}))
        if distances:
            indexofzero = distances.index(min(distances))
            distances.pop(indexofzero)
            dist_dict.pop(indexofzero)
            realindex = distances.index(min(distances))
            min_distances.append(min(distances))
            min_dist_dict.append(dist_dict[realindex])
        maini += 1
min_actual_dict = dict(zip(min_distances, min_dist_dict))
```

```python
by_uname_key_val_dist = {}
for index in range(len(min_dist_dict)):
    uname = list(min_dist_dict[index][0].keys())[0]
    dist = min_distances[index]
    by_uname_key_val_dist[uname] = dist
```
